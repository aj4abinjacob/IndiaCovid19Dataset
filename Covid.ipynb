```python
import pandas as pd
from datetime import datetime
from datetime import timedelta
```

```python
today = datetime.now()
yesterday = today - timedelta(days=1)
td = today.strftime('%d')
tm = today.strftime('%m')
ty = today.year
yd = yesterday.strftime('%d')
ym = yesterday.strftime('%m')
yy = yesterday.year
today = f'{tm}-{td}-{ty}'
yesterday = f'{yy}-{ym}-{yd}'
print('Today : ',today)
print('Yesterday : ',yesterday)
```

```python
url = f'https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_daily_reports/{today}.csv'
```

```python
ogdata = pd.read_csv('IndiaCovid19Dataset.csv')
ogdata.tail(25)
```

```python
df = pd.read_html(url)
df = df[0][1].str.split(',',expand = True)
df = df[df[3] == 'India']
df = df[[2,4,7,8,9,10]]
df.columns  = ogdata.columns
df.Date = yesterday
df = df[df['State'] != 'Unknown']
df
```

```python
ogdata = pd.concat([ogdata,df])
ogdata.tail(25)
```

```python
ogdata.to_csv('IndiaCovid19Dataset.csv',index=False)
ogdata.to_excel('IndiaCovid19Dataset.xlsx',index=False)
```
